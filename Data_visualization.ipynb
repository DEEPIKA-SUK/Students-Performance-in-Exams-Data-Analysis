**Students Performance in Exams**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
labels=["Female","Male"]
f_count=0
m_count=0
for i in df['gender']:
  if i=="female":
    f_count+=1
  else:
    m_count+=1
gen=[f_count,m_count]
plt.pie(gen,autopct='%.2f',labels=labels,shadow="true")
plt.legend(title = "Gender")
plt.show()
```

```python
col = "race/ethnicity"
grouped = df[col].value_counts().reset_index()
grouped = grouped.rename(columns = {col : "count", "index" : col})
race=df[col].unique()
data=grouped['count'].values
plt.pie(data,autopct='%.2f',labels=race)
plt.title("race/ethnicity")
plt.show()
```

```python
col = "parental level of education"
grouped = df[col].value_counts().reset_index()
grouped = grouped.rename(columns = {col : "count", "index" : col})
race=df[col].unique()
data=grouped['count'].values
plt.pie(data,autopct='%.2f',labels=race)
plt.title("parental level of education")
plt.show()
```

```python
col = "lunch"
grouped = df[col].value_counts().reset_index()
grouped = grouped.rename(columns = {col : "count", "index" : col})
race=df[col].unique()
data=grouped['count'].values
plt.pie(data,autopct='%.2f',labels=race)
plt.title("lunch")
plt.show()
```

```python
col = "test preparation course"
grouped = df[col].value_counts().reset_index()
grouped = grouped.rename(columns = {col : "count", "index" : col})
race=df[col].unique()
data=grouped['count'].values
plt.pie(data,autopct='%.2f',labels=race)
plt.title("test preparation course")
plt.show()
```
